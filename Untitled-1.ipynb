```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

```python
# URL of the CSV file
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Read CSV from the URL directly
df = pd.read_csv(url)

# Display the first rows to verify
print(df.head())

print('Data downloaded and read into a dataframe!')
```

### Develop a Line chart using the functionality of pandas to show how automobile sales fluctuate from year to year 

```python
# Group by Year and sum Automobile Sales
yearly_sales = df.groupby("Year")["Automobile_Sales"].sum()

# Plot
plt.figure(figsize=(10,6))
yearly_sales.plot(kind="line", marker="o")

plt.title("Total Automobile Sales per Year")
plt.xlabel("Year")
plt.ylabel("Total Automobile Sales")
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

### Annotations for at least 2 recession years

```python
# Group by Year
yearly_sales = df.groupby("Year")["Automobile_Sales"].sum()

# Plot
plt.figure(figsize=(12, 6))
ax = yearly_sales.plot(kind="line", marker="o", color="blue")

# Custom title
plt.title("Automobile Sales during Recession", fontsize=16)

# X-axis label
plt.xlabel("Year")
plt.ylabel("Total Automobile Sales")

# Ticks for all years
plt.xticks(yearly_sales.index, rotation=45)

# Grid
plt.grid(True)

# Annotations for at least 2 recession years
# For example: 2008 and 2020
if 2008 in yearly_sales.index:
    plt.annotate("Recession 2008",
                 xy=(2008, yearly_sales.loc[2008]),
                 xytext=(2008, yearly_sales.max()*0.9),
                 arrowprops=dict(facecolor='red', shrink=0.05),
                 fontsize=10,
                 color="red")

if 2020 in yearly_sales.index:
    plt.annotate("Covid-19 Recession",
                 xy=(2020, yearly_sales.loc[2020]),
                 xytext=(2020, yearly_sales.max()*0.7),
                 arrowprops=dict(facecolor='red', shrink=0.05),
                 fontsize=10,
                 color="red")

plt.tight_layout()
plt.show()
```

### Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?

```python
# Group by Year and Vehicle Type
vehicle_sales = df.groupby(["Year", "Vehicle_Type"])["Automobile_Sales"].sum().reset_index()

# Pivot so each Vehicle Type becomes a column
pivot_sales = vehicle_sales.pivot(index="Year", columns="Vehicle_Type", values="Automobile_Sales")

# Plot
plt.figure(figsize=(14, 7))

# Plot each line
for col in pivot_sales.columns:
    plt.plot(pivot_sales.index, pivot_sales[col], marker="o", label=col)

# Title and labels
plt.title("Automobile Sales per Vehicle Type Over Time", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Total Automobile Sales")
plt.xticks(pivot_sales.index, rotation=45)
plt.grid(True)
plt.legend(title="Vehicle Type")
plt.tight_layout()

plt.show()
```

### Insights on Sales Trends by Vehicle Type:
#### Overall Decline During Recessions:
##### All vehicle types show noticeable declines in sales during known recession years (e.g., 2008–2009 and 2020), confirming that recessions negatively impact sales across the board.
####  Variation in Sensitivity:
##### Executive Cars and Sports Cars exhibit steeper drops during recession periods, suggesting these categories are more sensitive to economic downturns (likely due to their higher prices and discretionary nature).
##### Small Family Cars and Medium Family Cars tend to maintain relatively higher volumes, even in recession years, indicating more stable demand.
#### Recovery Patterns:
##### After recessions, most vehicle types show recovery in sales, though the pace differs. For example, Sports Cars take longer to rebound, whereas Small Family Cars recover more quickly
#### Seasonal and Cyclical Effects:
##### There are some cyclical peaks and troughs across years, suggesting seasonality and broader economic cycles influence buying behavior.

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.

```python
# Group by Year and Recession flag, then calculate mean GDP
gdp_by_year = df.groupby(['Year', 'Recession'])['GDP'].mean().reset_index()

# Separate aggregated data
gdp_recession = gdp_by_year[gdp_by_year['Recession'] == 1]
gdp_non_recession = gdp_by_year[gdp_by_year['Recession'] == 0]

# Create figure and subplots
fig = plt.figure(figsize=(14, 6))

# Recession GDP
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(
    gdp_recession['Year'],
    gdp_recession['GDP'],
    color='red',
    marker='o',
    linestyle='-'
)
ax1.set_title('GDP During Recession')
ax1.set_xlabel('Year')
ax1.set_ylabel('Average GDP')
ax1.grid(True)

# Non-Recession GDP
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(
    gdp_non_recession['Year'],
    gdp_non_recession['GDP'],
    color='green',
    marker='o',
    linestyle='-'
)
ax2.set_title('GDP During Non-Recession')
ax2.set_xlabel('Year')
ax2.set_ylabel('Average GDP')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

### Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.

```python
# Filter non-recession data
df_non_recession = df[df["Recession"] == 0].copy()

# Convert Month to ordered categorical so it appears in calendar order
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
df_non_recession["Month_Name"] = pd.to_datetime(df_non_recession["Date"]).dt.month_name()
df_non_recession["Month_Name"] = pd.Categorical(df_non_recession["Month_Name"], categories=month_order, ordered=True)

# Create figure
plt.figure(figsize=(14,7))

# Bubble plot
plt.scatter(
    df_non_recession["Month_Name"],
    df_non_recession["Automobile_Sales"],
    s=df_non_recession["Seasonality_Weight"] * 500,  # scale bubble sizes
    alpha=0.6,
    edgecolors="w",
    color="skyblue"
)
# Titles and labels
plt.title("Seasonality Impact on Automobile Sales")
plt.xlabel("Month")
plt.ylabel("Automobile Sales")
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()
```

### Pie Chart of Advertising Expenditure by Recession Status

```python
# Group by Recession status and sum Advertising Expenditure
ad_expenditure = df.groupby("Recession")["Advertising_Expenditure"].sum()

# Labels
labels = ["Non-Recession", "Recession"]

# Colors
colors = ["#66c2a5", "#fc8d62"]

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    ad_expenditure,
    labels=labels,
    autopct="%1.1f%%",
    startangle=140,
    colors=colors,
    explode=(0, 0.1),
    shadow=True
)

# Title
plt.title("Advertising Expenditure During Recession and Non-Recession Periods")
plt.tight_layout()
plt.show()
```

###### Recession periods:
###### The slice representing recession spending is significantly smaller, showing that XYZAutomotives reduced advertising budgets during recessions. This is a common strategy as firms try to cut costs when consumer spending slows.
